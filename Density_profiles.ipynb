```python
import sys
import matplotlib.pyplot as plt

sys.path.append('/mnt/meom/workdir/martiene/scripts')
from Density_profiles import profile
```

The following class is defined with different arguments to compute density profile. Then use to ways to discretize the profile. The grad splits the layers using the highest values of the potential density gradient. Maximum search for the configuration where the p-norm of the layer mean density difference is bigger, i.e. maximizes $\left(\sum_i\left(\rho_{i+1}-\rho_i\right)^p\right)^{1/p}$, default $p=2$. For both configurations is possible to specify the minimum depth of the first layer and the maximum depth of the penultimate layer (both in meters), using variable _dep_. As well as the minimum layer thinckness (in number of layers of the imput data), using variable _nl_.

```python
GN = profile('GULF', '20090901')
GM = profile('GULF', '20100301')
ON = profile('OSMO', '20090901')
OM = profile('OSMO', '20100301')
```

```python
fig = plt.figure(figsize=[14, 6])
plt.subplot(121)
GN.plot_pdensity(label="November 1st", color='lightblue')
GN.make_partition(met="max", npar=2,  dep=(500, 1300), nlsep=20, color='blue', ls=':', label='max', save="GN3Lmax")
GN.make_partition(met="grad", npar=2,  dep=(400, 4000), nlsep=20,  color='blue', ls='-.', label='grad', save="GN3Lgra")
GM.plot_pdensity(label="March 1st", color='orange')
GM.make_partition(met="max", npar=2,  dep=(500, 1300), nlsep=20, color='brown', ls=':', label='max', save="GM3Lmax")
GM.make_partition(met="grad", npar=2,  dep=(400, 4000), nlsep=20,  color='brown', ls='-.', label='grad', save="GM3Lgra")
plt.legend()
plt.title("Gulf stream density profile")

plt.subplot(122)
ON.plot_pdensity(label="November 1st", color='tomato')
ON.make_partition(met="max", npar=2,  dep=(400, 1300), nlsep=20, color='red', ls=':', label='max')
ON.make_partition(met="grad", npar=2,  dep=(400, 4000), nlsep=20,  color='red', ls='-.', label='grad')
OM.plot_pdensity(label="March 1st", color='lightgreen')
OM.make_partition(met="max", npar=2,  dep=(400, 1300), nlsep=20, color='green', ls=':', label='max')
OM.make_partition(met="grad", npar=2,  dep=(400, 4000), nlsep=20,  color='green', ls='-.', label='grad')
plt.legend()
plt.title("Osmosis density profile")

plt.show()
plt.close()
```
